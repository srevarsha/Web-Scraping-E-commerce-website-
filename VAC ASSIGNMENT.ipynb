```python
import requests
```

```python
from  bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/beauty-and-grooming/makeup/eye-makeup/mascara/pr?sid=g9b,ffi,ttr,xbe&otracker=categorytree"
```

```python
allow=requests.get(url)
```

```python
allow
```

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

```python
product_name = soup.select(".s1Q9rs")
```

```python
product_name
```

```python
product_list =[]
```

```python
product_link =[]
```

```python
for item in product_name:
    product_list.append(item.text)
    link="https://www.flipkart.com/"
    product_link.append(link)
```

```python
product_list
```

```python
product_price = soup.select("._8VNy32 ._30jeq3")
```

```python
product_price
```

```python
price_list =[]
```

```python
for item in product_price:
    price_list.append(item.text)
```

```python
price_list
```
